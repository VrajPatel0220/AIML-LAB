```python
import pandas as pd
import numpy as np
```

```python
test = pd.read_csv('_test.csv')
train = pd.read_csv('_train.csv')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,f1_score,f1_score
from sklearn.preprocessing import OneHotEncoder
```

```python
print(train.isnull().sum())
print(test.isnull().sum())
```

```python
train['Age']=train['Age'].fillna(train['Age'].mean())
test['Age']=test['Age'].fillna(test['Age'].mean())
```

```python
train.dropna(subset='Embarked',axis=0,inplace=True) #drop rows where embarked is null
test.dropna(subset='Embarked',axis=0,inplace=True) #drop rows where embarked is null
```

```python
train.drop(columns=['Name','PassengerId','Cabin','Ticket'],inplace=True)
display(train.head())
test.drop(columns=['Name','PassengerId','Cabin','Ticket'],inplace=True)
display(test.head())
```

```python
test.dropna(subset='Fare',axis=0,inplace=True) #drop rows where fare is null
```

```python
print(train.isnull().sum())
print(test.isnull().sum())
```

```python
# onehot encoding
train = pd.get_dummies(train, columns=['Sex','Embarked'])
test = pd.get_dummies(test, columns=['Sex','Embarked'])
```
