```python

import pandas as pd
df = pd.read_csv('datasets/train.csv',usecols=['GarageQual','SalePrice','FireplaceQu'])
df.columns
```

```python
df.isnull().mean()*100
```

```python
df['GarageQual'].value_counts().plot(kind='bar')
```

```python
df['GarageQual'].mode()
```

```python
# Ideally frequent value imputation is used when one value occurs in majority
```

```python
df['GarageQual'].fillna('TA',inplace=True)
```

```python
df['GarageQual'].value_counts().plot(kind='bar')
```

```python
df['FireplaceQu'].value_counts().plot(kind='bar')
```

```python
import matplotlib.pyplot as plt

import seaborn as sns
sns.kdeplot(df[df['GarageQual'] == 'TA']['SalePrice'], shade=True)

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('KDE Plot Example')

# Show the plot
plt.show()
```
