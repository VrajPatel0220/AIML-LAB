```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv('../datasets/bread basket.csv')
df.head()
```

```python
df['Item'] = df['Item'].str.strip()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
grouped_df = df.groupby(['Transaction',"Item"])['Item'].count().reset_index(name='count')
grouped_df.head()
#grouping the data by transaction and item and counting the number of items in each transaction 
```

```python
basket_df = grouped_df.pivot_table(index='Transaction', columns='Item', values='count', aggfunc='sum').fillna(0)
#creating a pivot table with the transaction as the index and the items as the columns and the count of each item in the transaction as the values
```

```python
basket_df
```

```python
def hot_encode(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
```

```python
basket_encoded = basket_df.applymap(hot_encode)
basket_df=basket_encoded
#encoding the data to 1 and 0
#why do we need to encode the data if it already has 0 and 1 values ? because the data has the count of each item in the transaction and we need to convert it to 1 and 0
#why? because the apriori algorithm requires the data to be in the form of 1 and 0
basket_df
```

```python
frq_items = apriori(basket_df, min_support = 0.01, use_colnames = True)
# use col
frq_items
```

```python
rules = association_rules(frq_items, metric="confidence", min_threshold=1, support_only=False)
rules
```
